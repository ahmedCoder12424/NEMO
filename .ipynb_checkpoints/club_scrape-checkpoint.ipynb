```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service

import time
import random
```

```python
driver_path = 'chromedriver-win64/chromedriver.exe'
service = Service(driver_path)

driver = webdriver.Chrome(service=service)
```

```python
url = 'https://studentorg.ucsd.edu/'

driver.get(url)
```

```python
elements = driver.find_elements(By.XPATH, '//td')
clubs = []

for element in elements:
    clubs.append(element.text)
```

```python
elements = driver.find_elements(By.XPATH, '//td//a')
club_links = []

for element in elements:
    club_links.append(element.get_attribute('href'))
```

```python
instas = []
for link in club_links:
    driver.get(link)
    
    WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//dt[text()="Social Media"]')))
    dt_element = driver.find_element(By.XPATH, '//dt[text()="Social Media"]')
    insta = dt_element.find_element(By.XPATH, './following-sibling::dd[1]')
    
    if insta.text != '':
        instas.append(insta.text)
    
    # wait_time = random.uniform(1, 3)
    # time.sleep(wait_time)

instas
```

```python
def process(insta):
    if insta.startswith('Instagram: '):
        insta = insta[11:]
    if insta.startswith('instagram: '):
        insta = insta[11:]
    if insta.startswith('IG: '):
        insta = insta[4:]
    if insta.startswith('insta: '):
        insta = insta[7:]
    if insta.startswith('Insta: '):
        insta = insta[7:]
    if insta.startswith('Instagram '):
        insta = insta[10:]
    if insta.startswith('@'):
        insta = insta[1:]
    return insta
```

```python
for i in range(len(instas)):
    instas[i] = process(instas[i])

instas
```

```python
with open('club_instagrams.txt', 'w') as f:
    for line in instas:
        f.write(f"{line}\n")
```
